```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
from google.colab import drive
drive.mount('/content/drive')
```

##Image to Grayscale

```python
imgpath = '/content/drive/MyDrive/8th Term/CV-asg 1/Test images/6.jpg'
# imgpath = '/content/drive/MyDrive/computer vision/CV-asg 1/Test images/4.jpg'
```

```python
#gray_img = cv2.imread(, 0)
gray_img = cv2.imread(imgpath, 0)
print(gray_img.shape)
print(type(gray_img))
cv2_imshow(gray_img)
```

```python
print(gray_img)
```

##Noise Reduction Using Median Filter

```python
median = cv2.medianBlur(gray_img,9)
cv2_imshow(median)
```

##Edge Detection Using Laplacian Filter

```python
laplacian = cv2.Laplacian(median,ddepth=cv2.CV_8UC1,ksize=5)
print(laplacian.shape)
cv2_imshow(laplacian)
```

##Edges Thresholding

```python
Edgesimg = cv2.threshold(laplacian, 127, 255,cv2.THRESH_BINARY_INV)
cv2_imshow(Edgesimg[1])
print(Edgesimg[1])
```

#Generating a color painting and a cartoon

```python
img = cv2.imread(imgpath)
print(img.shape)
imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(imgRGB)
```

##Bilateral Filter

```python
blur = cv2.bilateralFilter(imgRGB,9,640, 426)
plt.imshow(blur)
```

##Creating Cartoon Effect

```python
cartoon_img = np.array(blur)
indexes = np.argwhere(Edgesimg[1] == 0)
for i in indexes:
  cartoon_img[i[0]][i[1]] = [0,0,0]
```

```python
plt.imshow(cartoon_img)
```
